Task 9

Display the names of the top 25 production companies based on the number of movies they have produced in descending order to the number of movies produced

```python
#Create an empty set
companies=set()

#Use a double for loop inside 'production companies' in order to split the production companies individually
for row in df_full["production_companies"].apply(lambda x: eval(x)):   #Lambda x:eval(x) evaluates the compiled set of strings

    for production_company in row:
        companies.add(production_company["name"]) #adds the name of the production company to the empty set
        
companies
```

```python
#Create empty list as to start a count of the production companies
bag_of_companies=""
for production_company in df_full["production_companies"]:
    bag_of_companies+=production_company
    
#Create an empty dictionary to store the companies and their respective counts
output=dict()
for company in companies:
    output[company]=bag_of_companies.count(company)
    
#Create a dataframe for the top 25 entries. Additionally, create a title for the column
result=pd.DataFrame()
for i in sorted(output.items(),key=lambda x:x[1],reverse=True)[0:25]:
    result.loc[i[0],"Count of Movies"]=i[1] #Creates the 'count of movies' column
                                            #Uses 'i' to insert the second index, which is the count of movies that has been 
                                            #placed in the output
    
result
```

Task 10

Sort the data in decending order based on revenue and filter the top 500 movies. Find the measures of central tendency for the following columns using the filtered data: Budget, Revenue, Runtime
    
Perform outlier analysis for the above three using box plots

```python
#Sort movies in decending order based on their revenue. Only include the top 500 entries
#Include the other two required measurements as well. 
rev_sort500=df_full.sort_values(by=['revenue'],ascending=[False]).head(500)[["budget","revenue","runtime"]]

#Creaete an empty dataframe
output=pd.DataFrame()

#Creates a new column called 'measure'. Additionally adds the three required columns stored in rev_sort500
output.loc[:,"Measure"]=rev_sort500.columns

#Sets up to use as a loop
c=["budget","revenue","runtime"]

for i in range(0,3):
    measure=c[i]
    
    #Set up all the variables required for outlier calculations
    Min=rev_sort500[measure].min()
    Max=rev_sort500[measure].max()
    Median=rev_sort500[measure].median()
    Range=rev_sort500[measure].max()-rev_sort500[measure].min()
    Q1=rev_sort500[measure].quantile(0.25)
    Q3=rev_sort500[measure].quantile(0.75)
    IQR=Q3-Q1
    Lower_Fence=Q1-1.5*IQR
    Upper_Fence=Q3+1.5*IQR
    
    #Insert the values found into the columns
    output.loc[i,'Min']=Min
    output.loc[i,'Max']=Max
    output.loc[i,'Median']=Median
    output.loc[i,'Range']=Range
    output.loc[i,'Q1']=Q1
    output.loc[i,'Q3']=Q3
    output.loc[i,'IQR']=IQR
    output.loc[i,'Lower_Fence']=Lower_Fence
    output.loc[i,'Upper_Fence']=Upper_Fence
    output.loc[i,'Outliers']=str(list(rev_sort500[(rev_sort500[measure]>=Upper_Fence) | (rev_sort500[measure]<=Lower_Fence)][measure].values))

#Changes format as to avoid scientific notation based format
pd.options.display.float_format = '{:.2f}'.format

output
```

```python
plt.figure(figsize=(12,6))
plt.subplots_adjust(wspace=0.5)

plt.subplot(131)
sns.boxplot(data=rev_sort500, y='budget')
plt.title('Budget')

plt.subplot(132)
sns.boxplot(data=rev_sort500, y='revenue')
plt.title('Revenue')

plt.subplot(133)
sns.boxplot(data=rev_sort500, y='runtime')
plt.title('Runtime')

plt.show()
```
